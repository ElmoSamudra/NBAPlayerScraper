```python
def get_per_game(player_name):
    
    '''searching a player's name and returning per game average summary statistics each year in the form of a nested list'''
    
    #search for player from the search bar
    driver.get("https://www.basketball-reference.com/")
    searchbox = driver.find_element_by_xpath('//*[@id="header"]/div[3]/form/div/div/input[2]')
    searchbox.click()
    searchbox.send_keys(player_name + "\n")
    
    #wait for page to completely load
    timeout = 5
    
    #get per_game table
    table = driver.find_element_by_xpath('//*[@id="per_game"]/tbody')
    
    #convert to list
    per_game=[]
    for tr in table.find_elements_by_xpath(".//tr"):
        data = tr.text.split()
        per_game.append(data)
        
    return per_game
```
